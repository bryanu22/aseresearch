```python
%matplotlib inline
%pip install statsmodels
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
# Load each city's data into a dictionary
data = {
	'Austin': pd.read_csv('Austin.csv', on_bad_lines='skip'),
	'Houston': pd.read_csv('Houston.csv', on_bad_lines='skip'),
	'El Paso': pd.read_csv('El Paso.csv', on_bad_lines='skip'),
	'Dallas': pd.read_csv('Dallas.csv', on_bad_lines='skip'),
	'San Antonio': pd.read_csv('San Antonio.csv', on_bad_lines='skip')
}
# Combine all city data into a single DataFrame with an additional 'City' column
combined_data = pd.concat(
	[df.assign(City=city) for city, df in data.items()],
	ignore_index=True
)

# Ensure the column contains numeric data
combined_data['Annual Summaries for Precipitation (2024), Stn: US1TXTV0262'] = pd.to_numeric(
	combined_data['Annual Summaries for Precipitation (2024), Stn: US1TXTV0262'], errors='coerce'
)

# Create a boxplot using a valid column name
combined_data.boxplot(column='Annual Summaries for Precipitation (2024), Stn: US1TXTV0262', by='City', figsize=(10, 6))
```
